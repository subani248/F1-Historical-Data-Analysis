```python
# Columns containing finishing positions
position_columns = [str(i) for i in range(1, 36)]

# Convert from wide format to long format
long_df = df.melt(
    id_vars=['Race', 'Date', 'Year', 'Month', 'Day'],
    value_vars=position_columns,
    var_name='Position',
    value_name='Driver'
)

long_df.head()
```

```python
long_df = long_df.dropna(subset=['Driver'])

long_df.head()
```

```python
long_df.info()
```

```python
long_df.shape
```

```python
long_df['Position'] = long_df['Position'].astype(int)

long_df.info()
```

```python
driver_races = long_df['Driver'].value_counts()

driver_races.head(10)
```

```python
plt.figure(figsize=(12,6))

driver_races.head(10).plot(kind='bar')

plt.title("Top 10 Drivers by Number of Race Finishes")
plt.xlabel("Driver")
plt.ylabel("Races Finished")

plt.xticks(rotation=45)

plt.show()
```

```python
average_finish = (
    long_df.groupby('Driver')['Position']
           .mean()
           .sort_values()
)

average_finish.head(10)
```

```python
driver_stats = long_df.groupby('Driver').agg(
    Total_Races=('Position', 'count'),
    Average_Position=('Position', 'mean')
)

driver_stats = driver_stats[driver_stats['Total_Races'] >= 50]

driver_stats = driver_stats.sort_values('Average_Position')

driver_stats.head(10)
```

```python
top_avg = driver_stats.head(10)

plt.figure(figsize=(12,6))

plt.bar(top_avg.index, top_avg['Average_Position'])

plt.title("Top 10 Drivers by Average Finishing Position (Min 50 Races)")
plt.xlabel("Driver")
plt.ylabel("Average Finishing Position")

plt.xticks(rotation=45)

plt.show()
```

```python
top10 = long_df[long_df['Position'] <= 10]

top10_counts = top10['Driver'].value_counts()

top10_counts.head(10)
```

```python
plt.figure(figsize=(12,6))

top10_counts.head(10).plot(kind='bar')

plt.title("Top 10 Drivers by Top-10 Finishes")
plt.xlabel("Driver")
plt.ylabel("Top-10 Finishes")

plt.xticks(rotation=45)

plt.show()
```

```python
# Total races
total_races = long_df.groupby('Driver').size()

# Wins
wins = long_df[long_df['Position'] == 1].groupby('Driver').size()

# Podiums
podiums = long_df[long_df['Position'] <= 3].groupby('Driver').size()

# Top 10
top10 = long_df[long_df['Position'] <= 10].groupby('Driver').size()

# Average Finish
avg_finish = long_df.groupby('Driver')['Position'].mean()
```

```python
driver_stats = pd.DataFrame({
    'Total Races': total_races,
    'Wins': wins,
    'Podiums': podiums,
    'Top 10 Finishes': top10,
    'Average Finish': avg_finish
})

driver_stats = driver_stats.fillna(0)

driver_stats['Wins'] = driver_stats['Wins'].astype(int)
driver_stats['Podiums'] = driver_stats['Podiums'].astype(int)
driver_stats['Top 10 Finishes'] = driver_stats['Top 10 Finishes'].astype(int)

driver_stats.head()
```

```python
driver_stats['Win Percentage'] = (
    driver_stats['Wins'] /
    driver_stats['Total Races']
) * 100

driver_stats.head()
```

```python
driver_stats = driver_stats.sort_values(
    by='Wins',
    ascending=False
)

driver_stats.head(10)
```

```python
driver_stats.to_csv("Driver_Statistics.csv")
```

```python
best_percentage = driver_stats[
    driver_stats['Total Races'] >= 30
].sort_values(
    by='Win Percentage',
    ascending=False
)

best_percentage.head(10)
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    best_percentage.head(10).index,
    best_percentage.head(10)['Win Percentage']
)

plt.title("Top 10 Drivers by Win Percentage")
plt.xlabel("Driver")
plt.ylabel("Win Percentage (%)")

plt.xticks(rotation=45)

plt.show()
```

# Final Project Conclusions

## Key Findings

- The analysis revealed a significant increase in the number of Formula 1 races over the years, demonstrating the continuous growth and global expansion of the championship.

- A small group of elite drivers has dominated the sport by securing a large proportion of total race victories, highlighting their exceptional performance and consistency.

- Drivers with a higher number of podium finishes generally achieved more race wins, indicating a strong positive relationship between consistent top-three finishes and overall success.

- The most successful drivers maintained lower average finishing positions throughout their careers, reflecting consistent high-level performance across multiple seasons.

- Race scheduling analysis showed that the majority of Formula 1 events take place between **May and October**, making it the busiest period of the racing calendar.

- Modern Formula 1 seasons include considerably more races than earlier decades, illustrating the expansion of the championship into new countries and racing circuits.
